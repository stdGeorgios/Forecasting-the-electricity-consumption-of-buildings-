```python
import os
import pandas as pd
from glob import glob

# ----------- Ρυθμίσεις ----------
base_path = r'C:\Plegma_Programming\Clean_Dataset_Final'
output_path = 'Processed_Houses_Final'
excluded_houses = {'House_06', 'House_08', 'House_12'}
os.makedirs(output_path, exist_ok=True)

# ----------- Συνάρτηση επεξεργασίας ανά σπίτι ----------
def process_house(house_path, house_id):
    print(f"\nΕπεξεργασία: {house_id}")
    
    # Φάκελοι
    electric_folder = os.path.join(house_path, 'Electric_data')
    env_folder = os.path.join(house_path, 'Environmental_data')

    # 1. Ανάγνωση αρχείων Electric_data
    electric_files = sorted(
        f for f in glob(os.path.join(electric_folder, '*.csv'))
        if 'appliances' not in os.path.basename(f).lower()
    )
    if not electric_files:
        print(f'[SKIPPED] {house_id}: Δεν βρέθηκαν αρχεία Electric_data')
        return

    df_elec = pd.concat([pd.read_csv(f, parse_dates=['timestamp']) for f in electric_files])
    df_elec.set_index('timestamp', inplace=True)
    df_elec = df_elec[df_elec['P_agg'].notna()]
    
    # Υπολογισμός ενέργειας ανά 30 λεπτά – Timestamp στο τέλος του διαστήματος
    energy_30min = df_elec['P_agg'].resample('30T', label='right', closed='right').sum() / 360
    df_energy = energy_30min.rename('energy_Wh').reset_index()

    # 2. Ανάγνωση αρχείων Environmental_data
    env_files = sorted(glob(os.path.join(env_folder, '*.csv')))
    if not env_files:
        print(f'[SKIPPED] {house_id}: Δεν βρέθηκαν αρχεία Environmental_data')
        return

    df_env = pd.concat([pd.read_csv(f, parse_dates=['timestamp']) for f in env_files])
    df_env.set_index('timestamp', inplace=True)
    df_env = df_env.interpolate(limit_direction='both')

    # Μέσος όρος κάθε 30 λεπτά με χρονική σφραγίδα στο τέλος του διαστήματος
    df_env_30min = df_env.resample('30T', label='right', closed='right').mean().reset_index()

    # 3. Συγχώνευση
    df_merged = pd.merge(df_energy, df_env_30min, on='timestamp', how='inner')

    # 4. Αποθήκευση
    out_file = os.path.join(output_path, f'{house_id}.csv')
    df_merged.to_csv(out_file, index=False)
    print(f'{house_id} → OK ({len(df_merged)} εγγραφές)')

# ----------- Εκτέλεση για όλα τα σπίτια ----------
all_houses = [d for d in os.listdir(base_path) if d.startswith('House_') and d not in excluded_houses]

for house_id in sorted(all_houses):
    house_path = os.path.join(base_path, house_id)
    process_house(house_path, house_id)

```
