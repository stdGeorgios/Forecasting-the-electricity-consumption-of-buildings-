```python
import os
import pandas as pd
from pathlib import Path

input_root = Path("C:/Plegma_Programming/Clean_Dataset")
output_root = Path("C:/Plegma_Programming/Clean_Dataset_Final")
excluded_houses = {"House_06", "House_08", "House_12"}

output_root.mkdir(exist_ok=True)

def fill_p_agg_with_similar_day(df):
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['weekday'] = df['timestamp'].dt.weekday
    df['hour'] = df['timestamp'].dt.hour
    df['minute'] = df['timestamp'].dt.minute

    known = df[~df['P_agg'].isna()]
    filled = df.copy()

    for idx in df[df['P_agg'].isna()].index:
        row = df.loc[idx]
        candidates = known[
            (known['weekday'] == row['weekday']) &
            (known['hour'] == row['hour']) &
            (known['minute'] == row['minute'])
        ]
        if not candidates.empty:
            filled.at[idx, 'P_agg'] = candidates['P_agg'].mean()
        else:
            filled.at[idx, 'P_agg'] = known['P_agg'].mean()
    return filled.drop(columns=["weekday", "hour", "minute"])

def process_house(house_path: Path, output_path: Path):
    electric_dir = house_path / "Electric_data"
    env_dir = house_path / "Environmental_data"

    new_electric_dir = output_path / "Electric_data"
    new_env_dir = output_path / "Environmental_data"
    new_electric_dir.mkdir(parents=True, exist_ok=True)
    new_env_dir.mkdir(parents=True, exist_ok=True)

    for file in sorted(electric_dir.glob("*.csv")):
        if file.name.lower().startswith("appliances_metadata"):
            continue
        df = pd.read_csv(file)
        if 'P_agg' in df.columns:
            df = fill_p_agg_with_similar_day(df)
        df.to_csv(new_electric_dir / file.name, index=False)

    for file in sorted(env_dir.glob("*.csv")):
        df = pd.read_csv(file)
        df = df.infer_objects(copy=False)
        df.interpolate(method='linear', inplace=True, limit_direction='both')
        df.to_csv(new_env_dir / file.name, index=False)

    print(f"[✓] Ολοκληρώθηκε για {house_path.name}")

for house_dir in sorted(input_root.iterdir()):
    if house_dir.is_dir() and house_dir.name not in excluded_houses:
        process_house(house_dir, output_root / house_dir.name)

print("✅ Ολοκληρώθηκε η προεπεξεργασία για όλα τα σπίτια.")
```
