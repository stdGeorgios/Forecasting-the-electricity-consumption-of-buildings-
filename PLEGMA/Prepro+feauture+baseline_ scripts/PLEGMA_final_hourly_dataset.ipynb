```python
import pandas as pd
from pathlib import Path

# =========================
# PATHS
# =========================
BASE_DIR = Path(r"C:\Plegma_Programming")

HOURLY_FILE = BASE_DIR / "PLEGMA_hourly_master_dataset.csv"
METADATA_FILE = BASE_DIR / "PLEGMA_metadata_filled.csv"

OUTPUT_FILE = BASE_DIR / "PLEGMA_final_hourly_dataset.csv"

# =========================
# LOAD DATA
# =========================
hourly_df = pd.read_csv(HOURLY_FILE)
meta_df = pd.read_csv(METADATA_FILE)

print("Hourly shape:", hourly_df.shape)
print("Metadata shape:", meta_df.shape)

# =========================
# MERGE
# =========================
df = pd.merge(hourly_df, meta_df, on="home_id", how="left")

print("\nAfter merge:", df.shape)

# =========================
# OPTIONAL: RENAME House -> home
# =========================
def convert_house_to_home(house_id):
    num = house_id.split("_")[1]
    return f"home{num}"

df["home_id"] = df["home_id"].apply(convert_house_to_home)

# =========================
# SORT
# =========================
df = df.sort_values(["home_id", "timestamp"]).reset_index(drop=True)

# =========================
# SAVE
# =========================
df.to_csv(OUTPUT_FILE, index=False)

print("\nSaved to:", OUTPUT_FILE)

# =========================
# PREVIEW
# =========================
print("\nColumns:")
print(df.columns.tolist())

print("\nSample:")
print(df.head(10))
```
