```python
import pandas as pd
from pathlib import Path

# =========================
# PATH
# =========================
FILE_PATH = Path(r"C:\Plegma_Programming\Processed_Houses_Final\House_01.csv")

# =========================
# LOAD
# =========================
df = pd.read_csv(FILE_PATH)

print("Before interpolation:")
print(df.isna().sum())

# =========================
# PREP
# =========================
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.sort_values("timestamp")

# =========================
# INTERPOLATE (μόνο environmental)
# =========================
cols_to_interp = [
    "internal_temperature",
    "external_temperature",
    "internal_humidity",
    "external_humidity"
]

df[cols_to_interp] = df[cols_to_interp].interpolate(method="linear")

# fallback για edge NaNs (αν υπάρχουν στην αρχή/τέλος)
df[cols_to_interp] = df[cols_to_interp].fillna(method="ffill").fillna(method="bfill")

# =========================
# CHECK
# =========================
print("\nAfter interpolation:")
print(df.isna().sum())

# =========================
# SAVE
# =========================
# επιλογή 1: overwrite
df.to_csv(FILE_PATH, index=False)

# επιλογή 2: αν θες backup version
# df.to_csv(FILE_PATH.parent / "House_01_fixed.csv", index=False)

print("\nFile updated successfully.")
```
