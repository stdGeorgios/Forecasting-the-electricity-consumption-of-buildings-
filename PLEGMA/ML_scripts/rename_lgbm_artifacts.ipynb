```python
from pathlib import Path
import shutil

# =========================
# CONFIG
# =========================
BASE_DIR = Path(r"C:/Plegma_Programming/models/final_lgbm")

# Αν θέλεις backup πριν rename
CREATE_BACKUP = True

# =========================
# FILE MAPPING
# =========================
rename_map = {
    "lgbm_cold.joblib": "lgbm_coldstart.joblib",
    "lgbm_history.joblib": "lgbm_withhistory.joblib",
    "lgbm_cold_features.json": "lgbm_coldstart_feature_columns.json",
    "lgbm_history_features.json": "lgbm_withhistory_feature_columns.json",
}

# =========================
# BACKUP (optional)
# =========================
if CREATE_BACKUP:
    backup_dir = BASE_DIR / "backup_before_rename"
    backup_dir.mkdir(exist_ok=True)

    print("\nCreating backup...")
    for old_name in rename_map.keys():
        src = BASE_DIR / old_name
        if src.exists():
            shutil.copy(src, backup_dir / old_name)
            print(f"Backed up: {old_name}")

# =========================
# RENAME
# =========================
print("\nRenaming files...\n")

for old_name, new_name in rename_map.items():
    old_path = BASE_DIR / old_name
    new_path = BASE_DIR / new_name

    if not old_path.exists():
        print(f"❌ Missing: {old_name}")
        continue

    if new_path.exists():
        print(f"⚠️ Already exists (skip): {new_name}")
        continue

    old_path.rename(new_path)
    print(f"✅ {old_name} → {new_name}")

print("\nDone.")
```
