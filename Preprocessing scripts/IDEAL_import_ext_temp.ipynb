```python
import pandas as pd
import requests

url = "https://archive-api.open-meteo.com/v1/archive"

def fetch_year(year, lat=55.9533, lon=-3.1883):
    params = {
        "latitude": lat,
        "longitude": lon,
        "start_date": f"2016-07-01",
        "end_date": f"2018-08-31",
        "hourly": "temperature_2m",
        "timezone": "Europe/London"
    }
    r = requests.get(url, params=params, timeout=60)
    r.raise_for_status()
    data = r.json()
    return pd.DataFrame({
        "timestamp": pd.to_datetime(data["hourly"]["time"]),
        "external_temperature": data["hourly"]["temperature_2m"]
    })

weather_all = pd.concat([fetch_year(y) for y in range(2016, 2018)], ignore_index=True)

print(weather_all.shape)
print(weather_all["external_temperature"].isna().sum())
weather_all.head()

weather_all["external_temperature"].describe()
weather_all["timestamp"].min(),
weather_all["timestamp"].max()

from pathlib import Path

WEATHER_DIR = Path("IDEAL_data/raw/weather")
WEATHER_DIR.mkdir(parents=True, exist_ok=True)

out_path = WEATHER_DIR / "IDEAL_edinburgh_hourly_temperature.csv"
weather_all.to_csv(out_path, index=False)

print("Saved to:", out_path.resolve())
```
