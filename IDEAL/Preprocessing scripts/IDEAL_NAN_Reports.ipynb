```python
import pandas as pd
from pathlib import Path

DATA_DIR = Path("E:/IDEAL/processed/hourly_per_home_final")

results = []

for csv_file in DATA_DIR.glob("home*_hourly.csv"):
    home_id = csv_file.stem.replace("_hourly", "")
    
    df = pd.read_csv(csv_file, parse_dates=["timestamp"])
    
    nan_counts = {
        "home_id": home_id,
        "nan_timestamp": df["timestamp"].isna().sum(),
        "nan_consumption": df["consumption_Wh"].isna().sum(),
        "nan_internal_temp": df["internal_temperature"].isna().sum(),
        "nan_external_temp": df["external_temperature"].isna().sum(),
        "total_rows": len(df)
    }
    
    results.append(nan_counts)

nan_report = pd.DataFrame(results)

nan_report.to_csv("E:/IDEAL/processed/nanreport.csv", index=False)

```
