```python
import pandas as pd
from pathlib import Path

BASE_DIR = Path("E:/IDEAL")

FINAL_DIR = BASE_DIR / "processed/hourly_per_home_final"
OUT_DIR = BASE_DIR / "processed"
OUT_DIR.mkdir(parents=True, exist_ok=True)

all_dfs = []

for csv_file in sorted(FINAL_DIR.glob("home*_hourly.csv")):
    home_id = csv_file.stem.replace("_hourly", "")
    df = pd.read_csv(csv_file, parse_dates=["timestamp"])
    df.insert(0, "home_id", home_id)
    all_dfs.append(df)

master = pd.concat(all_dfs, ignore_index=True)
master = master.sort_values(["home_id", "timestamp"]).reset_index(drop=True)

master_path = OUT_DIR / "IDEAL_master_hourly.csv"
master.to_csv(master_path, index=False)

print("Master dataset saved to:", master_path.resolve())
```
