```python
import pandas as pd
import numpy as np

# =========================================================
# SETTINGS
# =========================================================
DATA_PATH = r"C:\IDEAL_Programming\processed\final\IDEAL_final_hourly_dataset.csv"

HOME_ID = "home77"
TARGET_MONTH = 1
TARGET_DAY = 28

OUTPUT_PATH = r"C:\IDEAL_Programming\One_by_One\Home77\home77_real_profile_0128.csv"

# =========================================================
# LOAD DATA
# =========================================================
df = pd.read_csv(DATA_PATH)

df["timestamp"] = pd.to_datetime(df["timestamp"])

# =========================================================
# FILTER HOME
# =========================================================
df_home = df[df["home_id"] == HOME_ID].copy()

# =========================================================
# FILTER TARGET DATE (ALL YEARS)
# =========================================================
df_home["month"] = df_home["timestamp"].dt.month
df_home["day"] = df_home["timestamp"].dt.day
df_home["hour"] = df_home["timestamp"].dt.hour
df_home["date"] = df_home["timestamp"].dt.date

df_target = df_home[
    (df_home["month"] == TARGET_MONTH) &
    (df_home["day"] == TARGET_DAY)
].copy()

# =========================================================
# KEEP ONLY FULL DAYS (24 HOURS)
# =========================================================
counts = df_target.groupby("date")["hour"].count()
valid_dates = counts[counts == 24].index

df_target = df_target[df_target["date"].isin(valid_dates)]

if df_target.empty:
    raise ValueError("No full days found for this home and date.")

# =========================================================
# REAL PROFILE (MEAN PER HOUR)
# =========================================================
real_profile = (
    df_target.groupby("hour")["consumption_Wh"]
    .mean()
    .reset_index()
)

# rename for consistency
real_profile = real_profile.rename(columns={"consumption_Wh": "real_consumption_Wh"})

# =========================================================
# SAVE
# =========================================================
real_profile.to_csv(OUTPUT_PATH, index=False)

print("Saved real profile:", OUTPUT_PATH)
print(real_profile)
```
